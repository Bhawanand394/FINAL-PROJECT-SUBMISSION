```python
import matplotlib.pyplot as plt
```

```python
age=[58,69,43,39,63,52,47,31,74,36]
```

```python
chol=[189,235,193,177,154,191,213,165,198,181]
```

```python
n=len(age)
```

```python
sum_age=sum(age)
```

```python
sum_chol=sum(chol)
```

```python
mean_age=sum_age/n
```

```python
mean_chol=sum_chol/n
```

```python
sum_age2= sum(i**2 for i in age)
```

```python
sum_agechol=sum(age*chol for age,chol in zip(age,chol))
```

```python
print(sum_age)
```

```python
print(sum_chol)
```

```python
print(mean_age)
```

```python
print(mean_chol)
```

```python
beta_1=(n*sum_agechol)-(sum_age)*(sum_chol)
```

```python
beta_2=(n*sum_age2)-(sum_age)**2
```

```python
beta_3=beta_1/beta_2
```

```python
beta_0=mean_chol-beta_3*mean_age
```

```python
print("Slope",beta_3)
```

```python
print("Intercept",beta_0)
```

```python
predict_age=60
```

```python
predict_chol=beta_0+beta_3*predict_age
```

```python
print(predict_chol)
```

```python
plt.scatter(age,chol)
plt.xlabel("age")
plt.ylabel("chol")
plt.title("age vs chol")
plt.show()
```
